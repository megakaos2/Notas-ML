## Modelo logístico

Empezamos con un modelo básico, creamos un pipeline

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, scale, StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
# Extraer la variable endógena (y: churn)
y = df[['CHURN']]
#proc_data = df.drop(['CHURN'], axis=1)
#X = proc_data.values
X = df.drop(['CHURN'], axis=1)
```

Creamos un pipeline con dos pasos donde el primero normaliza los datos y el segundo aplica el modelo de regresión logística. Aprovechamos la función de stratify incluida en sklearn para balancear los datos.

```python
# pipeline = Pipeline([
#     ('scaler', StandardScaler(with_mean=False)),       # Step 1: Standardize features
#     ('reglog', LogisticRegression())  # Step 2: regresión logística
# ])
```

```python
## El df tiene 4022, así que con el 70% de los datos tenemos ~2800 para entrenar, preferimos dejar un 30% para probar la validación.

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, stratify=y)
```

Como los datos están desbalanceados procedemos a construir un pipeline para que normalice los datos y tomo los datos de una manera balanceada.

```python
# Logistic regression
from sklearn.linear_model import LogisticRegression
import time
from sklearn.pipeline import make_pipeline
import pickle

def useLogisticRegression(X_train,X_test,y_train,y_test):
    startTime =time.time()
    model = make_pipeline(StandardScaler(), LogisticRegression()) # Step 1: Standardize features, # Step 2: regresión logística
    model.fit(X_train, y_train)

    # Obtener el modelo de regresión logística dentro del pipeline
    log_model = model.named_steps["logisticregression"]

    feature_names = X_train.columns
    coefficients = pd.Series(log_model.coef_[0], index=feature_names)

    # Dibujar los coeficientes
    coefficients.sort_values().plot(kind='barh', figsize=(8, 6))
    plt.title("Feature Coefficients - Logistic Regression")
    plt.xlabel("Coefficient Value")
    plt.show()
    
    y_pred = model.predict(X_test)
    endTime = time.time()
    elapsed = endTime - startTime
    accuracy = accuracy_score(y_pred, y_test)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    reportSample = {
        'algorithm': 'Logistic',
        'timeTaken': elapsed,
        'accuracy': round(accuracy * 100, 3),
        'precision': round(precision, 3),
        'recall': round(recall, 3),
        'f1': f1_score(y_test, y_pred)
    }
    allReport.append(reportSample)
    accuracy = accuracy_score(y_pred, y_test)
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d')
    plt.xlabel('True Values negativo = 0 | positivo = 1')
    plt.ylabel('Predicted Values')
    plt.title('matriz de confusión Logistic Regression')
    plt.show()
    pickle.dump(model,open(f"models/{allModel.get('lr','model2')}",'wb'))

    return accuracy
```

```python
# Árbol de desiciones:
from sklearn import tree
import time
from sklearn.pipeline import make_pipeline
import pickle

def usedecisiontree(X_train,X_test,y_train,y_test):
    startTime =time.time()
    model = make_pipeline(StandardScaler(), tree.DecisionTreeClassifier()) # Step 1: Standardize features, # Step 2: regresión logística
    model.fit(X_train, y_train)   
    y_pred = model.predict(X_test)
    endTime = time.time()
    elapsed = endTime - startTime
    accuracy = accuracy_score(y_pred, y_test)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    reportSample = {
        'algorithm': 'Tree',
        'timeTaken': elapsed,
        'accuracy': round(accuracy * 100, 3),
        'precision': round(precision, 3),
        'recall': round(recall, 3),
        'f1': f1_score(y_test, y_pred)
    }
    allReport.append(reportSample)
    accuracy = accuracy_score(y_pred, y_test)
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d')
    plt.xlabel('True Values negativo = 0 | positivo = 1')
    plt.ylabel('Predicted Values')
    plt.title('matriz de confusión Desicion tree')
    plt.show()
    pickle.dump(model,open(f"models/{allModel.get('tree','model2')}",'wb'))

    return accuracy
```

```python
# modelo Naive Bayes:
from sklearn.naive_bayes import MultinomialNB
import time
from sklearn.pipeline import make_pipeline
import pickle

def useMultinominalNB(X_train,X_test,y_train,y_test):
    startTime =time.time()
    model = make_pipeline(MultinomialNB()) # Step 1: MultinomialNB
    model.fit(X_train, y_train)   
    y_pred = model.predict(X_test)
    endTime = time.time()
    elapsed = endTime - startTime
    accuracy = accuracy_score(y_pred, y_test)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    reportSample = {
        'algorithm': 'MultiNominal NaiveBayes',
        'timeTaken': elapsed,
        'accuracy': round(accuracy * 100, 3),
        'precision': round(precision, 3),
        'recall': round(recall, 3),
        'f1': f1_score(y_test, y_pred)
    }
    allReport.append(reportSample)
    accuracy = accuracy_score(y_pred, y_test)
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d')
    plt.xlabel('True Values negativo = 0 | positivo = 1')
    plt.ylabel('Predicted Values')
    plt.title('matriz de confusión MultiNominal NaiveBayes')
    plt.show()
    pickle.dump(model,open(f"models/{allModel.get('MultinomialNB','model2')}",'wb'))

    return accuracy
```

```python
# Modelo SVM
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV

def useSVCGridSearch(X_train, X_test, y_train, y_test):
    startTime = time.time()
    pipe = make_pipeline(StandardScaler(), SVC())
    param_grid = {
        'svc__degree': [2,3,4,5,6,7],
        'svc__gamma': [0.0001,0.001, 0.01, 0.1,1,10],
        'svc__C': [0.0001,0.001, 0.01, 0.1, 1,10]
    }

    grid = RandomizedSearchCV(pipe,param_distributions=param_grid,cv=5, n_iter=10, n_jobs=-1)
    grid.fit(X_train, y_train)
    model = grid.best_estimator_
    print(grid.best_score_)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    endTime = time.time()
    elapsed = endTime-startTime
    accuracy = accuracy_score(y_pred, y_test)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    reportSample = {
        'algorithm': 'SVM With Grid Search',
        'timeTaken': elapsed,
        'accuracy': round(accuracy * 100, 3),
        'precision': round(precision, 3),
        'recall': round(recall, 3),
        'f1': f1_score(y_test, y_pred)
    }
    allReport.append(reportSample)

    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d')
    plt.xlabel('True Values negativo = 0 | positivo = 1')
    plt.ylabel('Predicted Values')
    plt.title('matriz de confusión  SVC')
    plt.show()
    pickle.dump(model,open(f"models/{allModel.get('svc','model2')}",'wb'))
    return accuracy
```

```python
# Modelo Random Forest
from sklearn.ensemble import RandomForestClassifier

def useRandomGridSearch(X_train, X_test, y_train, y_test):
    startTime = time.time()
    pipe = make_pipeline(StandardScaler(), RandomForestClassifier())
    param_grid = {
        'randomforestclassifier__n_estimators': [10,20,30,40,50,100,150],
                 'randomforestclassifier__max_features': [5, 7, 9],
                 'randomforestclassifier__max_depth'   : [None, 3, 10, 20],
                 'randomforestclassifier__criterion'   : ['gini', 'entropy']
    }

    grid = RandomizedSearchCV(pipe,param_distributions=param_grid,cv=5, n_iter=10, n_jobs=-1)
    grid.fit(X_train, y_train)
    model = grid.best_estimator_
    print(grid.best_score_)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    endTime = time.time()
    elapsed = endTime-startTime
    accuracy = accuracy_score(y_pred, y_test)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    reportSample = {
        'algorithm': 'RandomForest With Grid',
        'timeTaken': elapsed,
        'accuracy': round(accuracy * 100, 3),
        'precision': round(precision, 3),
        'recall': round(recall, 3),
        'f1': f1_score(y_test, y_pred)
    }
    allReport.append(reportSample)

    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d')
    plt.xlabel('True Values negativo = 0 | positivo = 1')
    plt.ylabel('Predicted Values')
    plt.title('matriz de confusión Random Forest')
    plt.show()
    pickle.dump(model,open(f"models/{allModel.get('rf','model2')}",'wb'))
    return accuracy
```

Según el modelo de regresión logística, las variables que más influyen en el retiro de los clientes es el número de llamadas al servicio al cliente y si cuentan con plan internacional (positivo); de manera negativa o los que menos probabilidad tienen de retirarse serían los que cuentan con plan de voz. Tiene sentido pero el modelo logístico no arrojó buenos resultados.
