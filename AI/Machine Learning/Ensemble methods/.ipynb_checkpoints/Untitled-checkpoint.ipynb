```python
def Octal_c(n):
    res=[]
    if (n/8) > 1:
        while int(n/8)>0:
            res.append(n%8)
            n=int(n/8)
            if int(n/8) == 0:
                res.append(n%8)
    elif n/8==1:
        res.append(10)
    else:
        res.append(n%8)
    return ''.join(map(strres[::-1]))
```

```python
def Bin_c(n):
    res=[]
    if (n/2) > 1:
        while int(n/2)>=1:
            res.append(n%2)
            n=int(n/2)
            if int(n/2) == 0:
                res.append(1)
    elif n/2==1:
        res.append(10)
    else:
        res.append(n%2)
    return ''.join(map(strres[::-1]))
```

```python
def diagonalDifference(arr):
    diag1=[]
    for i in range(len(arr)):
        diag1.append(arr[i][i])

    sum_diag1=sum(diag1)
    
    diag2=[]
    for i in range(len(arr)):
        diag2.append(arr[i][(len(arr)-1)-i])
    sum_diag2=sum(diag2)

    return abs(sum_diag1-sum_diag2)
```

```python
def pangrams(s):
    s=s.lower()
    string_abc="abcdefghijklmnopqrstuvwxyz "
    comparison={x:0 for x in string_abc}
    for i in s:
        comparison[i]+=1

    if min(comparison.values())==0:
        return "not pangram"
    else:
        return "pangram"
    
```

```python
def flippingMatrix(matrix):
    return sum(
        max(matrix[i][j], matrix[~i][j], matrix[i][~j], matrix[~i][~j])
        for i in range(len(matrix) >> 1)
        for j in range(len(matrix) >> 1)
    )
```
