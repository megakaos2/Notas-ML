# Optuna

Es una librería que permite la optimización de cualquier función a partir de hiperparámetros:

Quick start:

```python
import optuna
optuna.logging.set_verbosity(optuna.logging.WARNING)

def objective(trial):
    x = trial.suggest_float('x', -2.1, 1)
    return 2*x**3+5*x**2-5*x+3

study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=1000, show_progress_bar=True, timeout=60*10)

#study.best_params  # E.g. {'x': 2.002108042}
print('Mejores hiperparámetros:', study.best_params)
```
