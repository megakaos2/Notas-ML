## Prediction

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('data/MBA.csv')
df["admission"]=df["admission"].replace("Admit",1.0).replace("Waitlist",2.0).fillna(0.0)
df["admission"]=pd.to_numeric(df["admission"])
```

```python
binary_encoder = OneHotEncoder(categories='auto', drop = 'first')
Company_1hot = binary_encoder.fit_transform(df[["gender","international","major","race","work_industry"]]).todense()
## Agregar el nombre de las columnas
column_name = binary_encoder.get_feature_names_out(["gender","international","major","race","work_industry"])
```

```python
onehot_df = pd.DataFrame(Company_1hot, columns=column_name)
```

```python
new_df=pd.concat([df[["gpa","gmat","work_exp","admission"]],onehot_df],axis=1,verify_integrity=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(new_df.drop(columns="admission"), new_df["admission"], test_size=0.2)

modelNB = MultinomialNB()
modelNB.fit(X_train,y_train)
```

```python
y_predict=modelNB.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score,confusion_matrix
from sklearn.metrics import accuracy_score

MSE = mean_squared_error(y_test,y_predict)
r2 = r2_score(y_test,y_predict)
cm = confusion_matrix(y_test,y_predict)

print(MSE,r2)
print("Train set Accuracy: ", accuracy_score(y_train, modelNB.predict(X_train)))
print("Test set Accuracy: ", accuracy_score(y_test, y_predict))
cm
```
