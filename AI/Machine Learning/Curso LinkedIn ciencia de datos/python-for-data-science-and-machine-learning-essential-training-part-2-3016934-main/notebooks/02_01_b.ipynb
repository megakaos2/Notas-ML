# Simple linear regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn

from pylab import rcParams
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 10,8
```

```python
rooms = 2*np.random.randn(100,1)+3
```

```python
# Se hace una transformación lineal a los números aleatorios para crear una relación lineal de la cual se estimarán los parámetros de esa recta

price = 265 + 6*rooms + abs(np.random.rand(100,1))
```

```python
plt.plot(rooms,price, 'r^')
plt.xlabel('# rooms')
plt.ylabel('2024 avg price')

## observar la relación (si es lineal o no)
```

```python
X = rooms
y = price

LinReg = LinearRegression()
LinReg.fit(X,y)
print(LinReg.intercept_, LinReg.coef_)
```

*Simple Algebra*
- y = mx + b
- b = intercept = 265.7

*Estimated Coefficients*
- LinReg.coef_ = [5.99] Estimated coefficients for the terms in the linear regression problem.

```python
print(LinReg.score(X,y))
```
