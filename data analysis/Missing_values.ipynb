```python
DF= pd.DataFrame(np.random.rand(36).reshape(6,6))
DF
```

```python
DF.loc[3:5,0] = np.nan
DF.loc[1:4,5] = np.nan
```

```python
# Se rellenan los na con 0
FilledDF=DF.fillna(0)
FilledDF
```

```python
# Se rellenan los na con valores personalizados de acuerdo al diccionario {columna 0:0.1, columna5:1.25}
FilledDF=DF.fillna({0:0.1,5:1.25})
FilledDF
```

```python
# Rellenar valores nulos con el Ãºltimo valor
fill_DF=DF.fillna(method="ffill")
fill_DF
```

```python
# Rellenar valores nulos con el siguiente valor
fill_DF=DF.fillna(method="bfill")
fill_DF
```
