# Spearman's Rank correlation

El supuesto asume que son variables ordinales numéricas, están relacionadas de manera no lineal y la data no está distribuida normal.

También se puede ver la independencia de las variables con un Chi-square test:
p < 0.05 rechaza la hipótesis nula (Hay correlación)
p > 0.05 acepta h0 (las variables son independientes)

```python
X = df[['cyl','vs','gear','am']]
```

```python
sb.pairplot(X)
```

```python
cyl=df['cyl']
vs=df['vs']
am=df['am']
gear=df['gear']

spearmanr_coef, pvalue = spearmanr(cyl,vs)
```

```python
print('Sp corr %0.3f' % (spearmanr_coef))
spearmanr_coef, pvalue = spearmanr(cyl,am)
print('Sp corr %0.3f' % (spearmanr_coef))
spearmanr_coef, pvalue = spearmanr(cyl,gear)
print('Sp corr %0.3f' % (spearmanr_coef))
```

```python
table = pd.crosstab(cyl,am)
```

```python
from scipy.stats import chi2_contingency
```

```python
chi2, p , dof, exp = chi2_contingency(table.values)
print('chi-sq %0.3f p_value %0.3f' % (chi2, p))
```

```python
table = pd.crosstab(cyl,vs)
chi2, p , dof, exp = chi2_contingency(table.values)
print('chi-sq %0.3f p_value %0.3f' % (chi2, p))
```

```python
table = pd.crosstab(cyl,gear)
chi2, p , dof, exp = chi2_contingency(table.values)
print('chi-sq %0.3f p_value %0.3f' % (chi2, p))
```
